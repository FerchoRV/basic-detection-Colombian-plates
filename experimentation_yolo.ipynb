```python
import easyocr
import cv2
import matplotlib.pyplot as plt

reader = easyocr.Reader(['es']) # 'es' para español/latino

img_placa = "runs/detect/results_folder/placas_1/crops/placa/Captura de pantalla 2026-04-16 200450.jpg"
img_placa_2 = "runs/detect/results_folder/placas_1/crops/placa/Captura de pantalla 2026-04-16 223753.jpg"
img_placa_3 = "runs/detect/results_folder/placas_1/crops/placa/Captura de pantalla 2026-04-16 223545.jpg"
img_placa_4 = "runs/detect/results_folder/placas_1/crops/placa/Captura de pantalla 2026-04-16 224011.jpg"
img = cv2.imread(img_placa_4)


resultado = reader.readtext(img)
placa=""
for (bbox, text, prob) in resultado:
    if prob > 0.5: # Filtrar por confianza
        #print(f"Texto detectado: {text.upper()} (Confianza: {prob:.2f})")
        placa = placa + text

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.title(f"Placa: {placa}")
plt.imshow(img_rgb)
plt.imshow(img_rgb)
plt.show()

```
